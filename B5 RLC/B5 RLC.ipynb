```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## RC

```python
rc = pd.read_csv('data\RC.csv')
rc['T'] = 1/rc['f']
rc['dphi'] = rc['dT']*2*np.pi/rc['T']
rc
```

```python
fig = plt.figure(figsize=(10, 5))
plt.plot(np.log10(rc['f']), rc['UC'], 'o-', color='lightseagreen', label='$U_C$')
plt.plot(np.log10(rc['f']), rc['UR'], 'o-', color='lightcoral', label='$U_R$')

plt.xlabel('$log_{10}$ Frequency $log_{10}f/log_{10}Hz$')
plt.ylabel('Voltage U/V')
plt.title('Fig. 1 RC Amplitude-frequency Curve', y=-0.3)
plt.grid()
plt.legend(loc='center left', bbox_to_anchor=(0.4, -0.185), ncol=2)
plt.savefig(r'output\fig. 1.png', bbox_inches = 'tight', dpi=300)
```

```python
fig = plt.figure(figsize=(10, 5))
plt.plot(np.log10(rc['f']), rc['dphi'], 'o-', color='lightseagreen')

plt.xlabel('$log_{10}$ Frequency $log_{10}f/log_{10}Hz$')
plt.ylabel('Phase difference $\Delta\phi/rad$')
plt.title('Fig. 2 RC Phase-frequency Curve', y=-0.2)
plt.grid()
plt.savefig(r'output\fig. 2.png', bbox_inches = 'tight', dpi=300)
```

## RLC

```python
RLC = pd.read_csv('data\RLC.csv')
RLC['T'] = 1/RLC['f']
RLC['dphi'] = RLC['dT']*2*np.pi/RLC['T']
RLC['I'] = RLC['UR']/510
RLC
```

```python
fig = plt.figure(figsize=(10, 5))
plt.plot(np.log10(RLC['f']), RLC['I'], 'o-', color='lightseagreen')

plt.xlabel('$log_{10}$ Frequency $log_{10}f/log_{10}Hz$')
plt.ylabel('Effective current $I/A$')
plt.title('Fig. 5 RLC Amplitude-frequency Curve', y=-0.2)
plt.axvline(x=np.log10(1.64e+3), ls=':', color='lightcoral')
plt.annotate(text='f=1.64kHz', color='lightcoral', xy=(np.log10(1.64e+3), 0.00293), xytext=(3.25, 0.0025), arrowprops=dict(arrowstyle='->'
                                ,color='lightcoral',
                                shrinkA = 5,
                                shrinkB = 5,
                                connectionstyle="arc3,rad=0.3"))
plt.grid()
plt.savefig(r'output\fig. 5.png', bbox_inches = 'tight', dpi=300)
```

```python
fig = plt.figure(figsize=(10, 5))
plt.plot(np.log10(RLC['f']), RLC['dphi'], 'o-', color='lightseagreen')

plt.xlabel('$log_{10}$ Frequency $log_{10}f/log_{10}Hz$')
plt.ylabel('Phase difference $\Delta\phi/rad$')
plt.title('Fig. 6 RLC Phase-frequency Curve', y=-0.2)
plt.axvline(x=np.log10(1.64e+3), ls=':', color='lightcoral')
plt.annotate(text='f=1.64kHz', color='lightcoral', xy=(np.log10(1.64e+3), 0), xytext=(3.5, .5), arrowprops=dict(arrowstyle='->'
                                ,color='lightcoral',
                                shrinkA = 5,
                                shrinkB = 5,
                                connectionstyle="arc3,rad=0.3"))
plt.grid()
plt.savefig(r'output\fig. 6.png', bbox_inches = 'tight', dpi=300)
```
